```python
#TASK 1
# Question 1 

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Question 2

# Load Iris dataset directly from seaborn
df = sns.load_dataset('iris')

# Question 3

# Display the first few rows
print(df.head())
```

```python
#Question 4

# Display data types and missing values
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#Question 5

# Cleaning of the dataset 
df = df.dropna()
```

```python
#TASK 2
#Question 1

# Basic statistics
print("\nBasic Statistics:\n", df.describe())
```

```python
#Qestion 2

# Grouping by species and calculating mean of numerical columns
grouped = df.groupby('species').mean()
print("\nMean by Species:\n", grouped)
```

```python
#Question 3

# Observations and some insiteful findings
for col in df.select_dtypes(include='float'):
    print(f"Mean {col} by species:\n{df.groupby('species')[col].mean()}\n")
```

```python
#TASK 3 DATA VISUALIZATION
#Question 1 Linechart

# Set seaborn style
sns.set(style="whitegrid")

# Line Chart: Dummy trend using index to simulate time (Iris has no date column)
df['index'] = df.index
plt.figure(figsize=(10, 4))
sns.lineplot(x='index', y='sepal_length', data=df)
plt.title("Simulated Sepal Length Trend Over Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
#Question 1 Bar chart

# Bar Chart: Average petal length per species
plt.figure(figsize=(6, 4))
sns.barplot(x='species', y='petal_length', data=df)
plt.title("Average Petal Length per Species")
plt.ylabel("Petal Length")
plt.xlabel("Species")
plt.show()
```

```python
#Question 1 Histogram

# Histogram: Sepal Width Distribution
plt.figure(figsize=(6, 4))
sns.histplot(df['sepal_width'], bins=15, kde=True)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.show()
```

```python
#Question 1 Scatter plot

# Scatter Plot: Sepal Length vs Petal Length
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend(title="Species")
plt.show()
```
